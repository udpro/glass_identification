# Glass Identification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
glass=pd.read_csv("glass.csv")
names = ['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','type']
glass.columns = names
```

```python
glass
```

```python
glass.plot()
```

```python
glass.plot(kind = 'hist', x='RI' , y='type')
plt.show()
```

```python
glass.plot(kind = 'scatter', x='Na' , y='type')
plt.show()
```

```python
features = ['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']
label = ['type']

X = glass[features]
Y = glass[label]
```

```python
x2=X.values

from matplotlib import pyplot as plt
import seaborn as sns
for i in range(1,9):
    sns.histplot(x2[i])
    plt.xlabel(features[i])
    plt.show()
```

```python
import seaborn as sns
x2 = pd.DataFrame(X)

plt.figure(figsize=(8,8))
sns.pairplot(data=x2)
plt.show()
```

```python
coreleation= X.corr()
plt.figure(figsize=(15,15))
sns.heatmap(coreleation,cbar=True,square=True,annot=True,fmt='.1f',annot_kws={'size': 15},xticklabels=features,yticklabels=features,alpha=0.7,cmap= 'coolwarm')
plt.show()
     
```
